```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering, KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
```

```python
data = pd.read_csv("absenteeism_0.csv")
```

```python
features = data.iloc[:, :-1]

# Normalize the features
scaler = StandardScaler()
features = scaler.fit_transform(features)

# Perform hierarchical clustering
hierarchical_cluster = AgglomerativeClustering(n_clusters=3)
hierarchical_labels = hierarchical_cluster.fit_predict(features)

# Perform K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(features)
```

```python
data['Hierarchical_Cluster'] = hierarchical_labels
data['KMeans_Cluster'] = kmeans_labels

# Calculate the centroids of K-means clusters
centroids = kmeans.cluster_centers_

# Plot the clusters
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Plot Hierarchical Clusters
colors = ['red', 'green', 'blue']
for i in range(3):
    cluster = data[data['Hierarchical_Cluster'] == i]
    ax1.scatter(cluster['Transportation_expense'], cluster['Distance_from_Residence_to_Work'], c=colors[i])
ax1.set_title("Hierarchical Clustering")

# Plot KMeans Clusters
colors = ['red', 'green', 'blue']
for i in range(3):
    cluster = data[data['KMeans_Cluster'] == i]
    ax2.scatter(cluster['Transportation_expense'], cluster['Distance_from_Residence_to_Work'], c=colors[i])
ax2.scatter(centroids[:, 0], centroids[:, 1], marker='*', s=200, c='#050505')
ax2.set_title("KMeans Clustering")

plt.show()
```
