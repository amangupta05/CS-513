```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Define the dataset
data = {'Occupation': ['Service', 'Management', 'Sales', 'Staff', 'Management', 'Sales', 'Management', 'Staff', 'Service', 'Management', 'Sales'],
        'Gender': ['Female', 'Male', 'Male', 'Male', 'Female', 'Male', 'Female', 'Female', 'Male', 'Female', 'Male'],
        'Age': [45, 25, 33, 25, 35, 26, 45, 40, 30, 50, 25],
        'Salary': ['Level 3', 'Level 1', 'Level 2', 'Level 3', 'Level 4', 'Level 3', 'Level 4', 'Level 3', 'Level 2', 'Level 2', 'Level 1']}

df = pd.DataFrame(data)

# One-hot encode the categorical variables
df = pd.get_dummies(df, columns=['Occupation', 'Gender'])

# Convert salary levels to numerical values
salary_map = {'Level 1': 1, 'Level 2': 2, 'Level 3': 3, 'Level 4': 4}
df['Salary'] = df['Salary'].map(salary_map)

# Define the feature matrix X and target variable y
X = df.drop('Salary', axis=1)
y = df['Salary']

# Fit a decision tree classifier with max_depth=1
tree = DecisionTreeClassifier(max_depth=1)
tree.fit(X, y)

# Plot the decision tree
plot_tree(tree, feature_names=X.columns)
```
